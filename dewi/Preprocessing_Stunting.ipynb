```python
# Baca data dari file (misalnya CSV)
# Membaca data dari file CSV
data = pd.read_csv('/content/drive/MyDrive/SIDANG SKRIPSI/DATA/data_pre.csv', sep=';')
data
```

```python
# Mendapatkan jumlah baris
jumlah_data = data.shape[0]

print(f"Jumlah data: {jumlah_data}")
```

```python
# Menggunakan method isna() untuk mendeteksi data yang kosong
missing_data = data.isna()

# Menggunakan method sum() untuk menghitung jumlah data yang kosong dalam setiap kolom
jumlah_data_kosong = missing_data.sum()

# Tampilkan hasil
print(jumlah_data_kosong)
```

```python
# Menghapus baris dengan nilai yang hilang (NaN)
data = data.dropna()

# Menghapus duplikat data
data = data.drop_duplicates()
```

### Prepocessing

```python
# Melakukan one-hot encoding pada kolom 'JK' dan mengubah True/False menjadi 1/0
data = pd.get_dummies(data, columns=['JK'], prefix='JK', dtype=int)

print(data)
```

```python
#Mengubah value usia menjadi sebuah angka
data['Usia Saat Ukur'] = data['Usia Saat Ukur'].apply(lambda x : float(x[0]))

data.head()
```

```python
# Menampilkan DataFrame sebelum penghapusan kolom
print("DataFrame sebelum penghapusan kolom:")
print(data)
```

```python
# Kolom-kolom yang akan dihapus
columns_to_drop = ['No', 'Nama', 'Tgl Lahir', 'Tanggal Pengukuran', 'bb/tb', 'z-core bb/tb', 'bb/u', 'z-core bb/u', 'tb/u', 'z-core tb/u']

# Menghapus kolom-kolom tertentu dari DataFrame
data = data.drop(columns=columns_to_drop)

# Menampilkan DataFrame setelah penghapusan kolom
print("\nDataFrame setelah penghapusan kolom:")
print(data)
```

```python
# Mengkodekan variabel kategorikal (misalnya, jenis kelamin) menjadi angka
data['Diagnosa'] = data['Diagnosa'].map({'TIDAK': 0, 'IYA': 1})
data
```

```python
# Memindahkan kolom 'Diagnosa' ke belakang
cols = [col for col in data.columns if col != 'Diagnosa'] + ['Diagnosa']
data = data[cols]

print(data)
```

```python
from sklearn.preprocessing import MinMaxScaler

# Inisialisasi MinMaxScaler
scaler = MinMaxScaler()

# Pilih kolom yang akan dinormalisasi (dalam hal ini, semua kolom)
columns_to_normalize = data.columns

# Lakukan normalisasi
data[columns_to_normalize] = scaler.fit_transform(data[columns_to_normalize])

# Tampilkan dataframe setelah normalisasi
data_normalized = data
print(data)
```
